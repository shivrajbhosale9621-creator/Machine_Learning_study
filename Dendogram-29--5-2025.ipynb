```python
import scipy.cluster.hierarchy as sch
val=[[2,5],[3,6],[7,10],[12,10],[6,3],[10,20]]
sch.linkage(val)
```

```python
import matplotlib.pyplot as plt
dendogram=sch.dendrogram(sch.linkage(val,method='single'))
plt.title('Dendrogram')
plt.xlabel('Coustomers')
plt.ylabel('Distance')
plt.show()
```

```python
import numpy as np
import pandas as pd
import numpy.linalg as limm
data=[]
for i in range(len(val)):
    for j in range(len(val)):
        p1=np.array(val[i])
        p2=np.array(val[j])
        #print(f'Distance between {p1} with Pp2})
        dis=np.linalg.norm(p1-p2)
        data.append([p1,p2,dis])
df=pd.DataFrame(data,columns=['p1','p2','dis'])
df.sort_values(by='dis').iloc[6:,:]
```
