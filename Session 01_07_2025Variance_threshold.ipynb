## Infromation gain

mutual info classify

- It is a value estimating how much relationship/how much information about target variable

- We dont want relation between input features

- We want relation between input and output feature

- 0 means both variables are independent

- Here will use entropy or KNN method based on joint probability

```python
from sklearn.feature_selection import mutual_info_classif
MI=mutual_info_classif(X,y)
```

```python
MI
```

```python
d=pd.DataFrame(MI,index=X.columns,columns=['Mutual_info_classif'])
```

```python
sorted_data=d.sort_values(by='Mutual_info_classif',ascending=False)
sorted_data
```

```python
con=sorted_data['Mutual_info_classif']>0.10
cols=sorted_data
```

```python
d=pd.DataFrame(MI,
              columns=['Information gain'],
              index=X.columns)
sorted_d=d.sort_values('Information gain',
                      ascending=False)
df[sorted_d.index[:5]]
```
