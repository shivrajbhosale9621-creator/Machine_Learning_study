```python
import pandas as pd
dict1={'A':[100,200,300,400],
      'B':[1,2,3,4],
      'C':[10,20,30,40],
      'D':[10,10,10,10]}
df=pd.DataFrame(dict1)
df
```

```python
from sklearn.feature_selection import VarianceThreshold

# Initialize the VarianceThreshold instance with a threshold of 0
vt = VarianceThreshold(threshold=0)

# Fit the model to the data (assuming df is your DataFrame)
vt.fit(df)

```

## Information gain

mutual info classify

- It is a valu estimating how much relationship/how much information about target variable

- we dont want relation between input features

- we want relation between input and output feture

- 0 means both variables are indepenent

- Here will use entropy or KNN method based on joint probability

- Mutual info classift will take input columns data and output colummns data

- wright it leter

```python
data=pd.read_csv(r"C:\Data Sets\winequality_red.csv")
x=data.drop('quality',axis=1) 
y=data['quality']
```

```python
len(data.columns),len(x.columns)
```

```python
x.columns
```

```python
from sklearn.feature_selection import mutual_info_classif
MI=mutual_info_classif(x,y)
MI
```

```python
d=pd.DataFrame(MI,index=x.columns,
              columns=['MI'])
```

```python
d
```

```python
sorted_data=d.sort_values(by="MI",ascending=False)
sorted_data
```
